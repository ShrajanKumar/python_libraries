### Adjusting the spacing around the subplots

By default matplotlib leaves a certain amount of padding around the outside of the
 subplots and spacing between subplots. 

You can
 change the spacing using the subplots_adjust method on Figure objects, also avail
able as a top-level function:

subplots_adjust(left=None, bottom=None, right=None, top=None,
                wspace=None, hspace=None)

```python
fig,axes=plt.subplots(2,2 ,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
plt.subplots_adjust(right=3,top=3,wspace=0.2,hspace=0.2)
```

```python
fig,axes=plt.subplots(2,2)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
plt.subplots_adjust(wspace=0,hspace=0)
```

## Ticks and titles

###### ticks labels titles and ticklabels

```python
plt.plot(np.random.randn(1000).cumsum())
```

This is the defaultt plot now let us define x axis ticks and labels title for graph and x axis labels

 To change the x-axis ticks, it’s easiest to use set_xticks and set_xticklabels. The
former instructs matplotlib where to place the ticks along the data range; by default
these locations will also be the labels. But we can set any other values as the labels
 using set_xticklabels.
 
 The rotation option sets the x tick labels at a 30-degree rotation. Lastly, set_xlabel
 gives a name to the x-axis and set_title the subplot title

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
ax.set_xticks([0,250,500,750])
ax.set_xticklabels(['one','two','three','four'],rotation=90,fontsize='medium')
ax.set_title('My first plot')
ax.set_xlabel('Stages')
```

###### legends

Legends are the important component of plots for identifying different plots in a single graph.Easisest way to add legend is to pass labels while adding each piece of graph and in end call .legend()

```python
from numpy.random import randn as rn

fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(rn(1000).cumsum(),'r',label='first')
ax.plot(rn(1000).cumsum(),'k--',label='second')
ax.plot(rn(1000).cumsum(),'g.',label='Third')
ax.legend(loc='best') #setting legend  loc is the location in graph to display legend

#for better understanding of the graph use below 
fig.subplots_adjust(right=1.5,top=1.5)
```

'upper left', 'upper right', 'lower left', 'lower right'  are  some of the other usefull values for loc (legend location)

## Plotting with  Pandas and Seaborn 

matplotlib can be a fairly low-level tool. You assemble a plot from its base components: the data display (i.e., the type of plot: line, bar, box, scatter, contour, etc.), legend, title, tick labels, and other annotations.

In pandas we may have multiple columns of data, along with row and column labels.

#### Line plot

```python
import pandas as pd
import numpy as np
s1=pd.Series(np.random.randn(10).cumsum(),index=np.arange(100,200,10))

s1.plot()
```

One can observe that series index is used as x-axis labels. By setting use_index=False one can skip using index as xlabel

```python
s1.plot(use_index=False,style='k--',alpha=0.5)
```

The x-axis ticks and limits can be
 adjusted with the xticks and xlim options,  similary can be done for y axis

```python
s1.plot(style='k--',xticks=[120,140,160],xlim=[100,180])
```

DataFrame’s plot method plots each of its columns as a different line on the same
 subplot, creating a legend automatically.

```python
df=pd.DataFrame(np.random.randn(10,4).cumsum(0),columns=['one','two','three','four'],
                index=np.arange(0,100,10))

df.plot()
```

### Barplot

plot.bar() and df.barh() make vertical and horizontal bar plots, respec
tively.  In bar() index will be used in x axis and in barh()  index will be used in y axis

```python
s1=pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))

fig,axes=plt.subplots(2,1)
s1.plot.bar(ax=axes[0],color='k',alpha=0.5)
s1.plot.barh(ax=axes[1],color='k',alpha=0.5,)
fig.subplots_adjust(top=2 ,hspace=0.1)
```

With a DataFrame, bar plots group the values in each row together in a group in bars,
 side by side, for each value. 

```python
df=pd.DataFrame(np.random.rand(6,4),index=('one two three four five six').split(),columns=list('abcd'))
df
```

```python
df.plot.bar()
```

One can also plot stacked bargraph as by passing stacked=True in side plot argument.



```python
df.plot.barh(stacked=True , alpha=0.6)
```

If you want to compare different columns within a row (xticks) dont stack but if u want to compare column value over different row itt is help full to use stack (it helps to reduce number of graph needed)

A useful recipe for bar plots is to visualize a Series’s value frequency 
using value_counts: s.value_counts().plot.bar().

```python
s1=pd.Series(['a','b','c','a','a','b','c','c','c','c'])
s1.value_counts().plot.bar(color='g')
plt.title('Frequency graph')
plt.xlabel('values')
plt.ylabel('Frequency')
```
